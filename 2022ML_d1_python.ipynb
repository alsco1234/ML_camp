```python
# Python Controls
# if / else if / else
a = 0

if a == 1:
  print("a == 1")
elif a == 0:
  print("a == 0")
else:
  print("a == somthing else")

# Nested if / else if / else
a = 9

if a > 0:
  if a % 3 == 0:
    if a % 9 == 0:
      print('a is positive and divisible by both 3 and 9')
    else :
      print('a is positive and divisible by 3')
  else :
    print('a is a positive number')
elif a == 0:
  print('a is a positive number')
else :
  print('a is a negative number')
```

```python
def sum_to_ten(n):
  if n < 1:
    return 0
  return sum_to_ten(n-1) + n

print('sum_to_ten(10) = ', sum_to_ten(10))
print('sum_to_ten(9) = ', sum_to_ten(9))
print('sum_to_ten(5) = ', sum_to_ten(15))
```

```python
# Q1
def right_triangle(n):
  for i in range(1,n+1,1):
    for j in range(1,i+1):
      print('#', end=' ')
    print('')

right_triangle(3)
```

```python
# Q2
def palidrome(str):
  boolres = False
  strres = ''

  for i in str:
    strres = i + strres
  
  if str == strres:
    boolres = True

  return boolres, strres

a, b = palidrome('apple')
print('returns a = ', a , ' and b = ', b)
c, d = palidrome('radar')
print('returns c = ', c , ' and d = ', d)
```
