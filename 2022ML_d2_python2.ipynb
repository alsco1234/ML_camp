```python
# 2. 파이썬 함수는 여러 개의 return 값을 가질 수 있다

def func():
  return 1, 2, 3, 4, 5

one, two, three, four, five = func()
print(one, two, three, four, five)

a, _, _, b, c = func() # _로 값을 받아 버림
print(a, b, c)
```

```python
# 3. for 끝날때 else 받아 마지막 조건을 걸 수 있다

def func(array):
  for num in array:
    if num%2 == 0:
      print(num)
      break
  else:
    print('No call for Break. Else is executed')
    
print("1st Case:")
a = [2]
func(a)
print("2nd Case:")
a = [1]
func(a)
```

```python
# 4. 여러개가 묶인 list나 dictionary을 풀 수 있다

def point(x, y):
  print(x, y)

foo_list = (3, 4)
bar_dict = {'y':3, 'x':2}

point(*foo_list)
point(**bar_dict) #hash_table은 순서가 보장되지 않는다

colors = ['red', 'green', 'blue', 'yellow', 'cyan', 'purple']
print(*colors)
```

```python
# 5. enumerate로 iterator 만들어 for에 편하게 이용할 수 있다 (자바처럼!)

vowels = ['a', 'e', 'i', 'o', 'u']
for i, letter in enumerate(vowels):
  print(i, letter)
```

```python
# 6. 비교연산자를 두개 같이 쓸 수 있다

i = 5;

ans = 1 < i < 10
print(ans)

ans = 10 > i <= 9
print(ans)

ans = 5 == i
print(ans)
```

```python
# 7. infinity(표현할 수 없는 최소/최대값)을 이용할 수 있다

# 1) positive
p_infinity = float('Inf')
if 9999999999999 > p_infinity:
  print("The number is greater than Infinity!")
else:
  print("Infinity is greatest")

# 2) negative
n_infinity = float('-Inf')
if -9999999999999 < n_infinity:
  print("The number is less than negative Infinity!")
else:
  print("Negative Infinity is least")

```

```python
# 8. list comprehension으로 쉽게 추가한다

# 1) append
a = []
for x in range(0, 10):
  a.append(x)
print(a)

# 2) comprehension
print([x for x in a])

```

```python
# 9. 다양한 index 슬라이싱

a = [1, 2, 3, 4, 5]

print(a[0:2]) # 1,2
print(a[0:-1]) # 1,2,3,4
print(a[::-1]) # 5,4,3,2,1
print(a[::2]) # 1,3,5
print(a[::-2]) # 5,3,1
```
